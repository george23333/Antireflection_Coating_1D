```python
def complex_abs2(z: torch.Tensor) -> torch.Tensor:
    return z.real.square() + z.imag.square()

z = torch.tensor([1 + 2j, 3 + 4j])
print(complex_abs2(z))
```

```python
import torch
import torch.autograd as autograd

x = torch.tensor([[2.0]], requires_grad=True)  # x = 2

u = x**2   # u = 4
```

```python
du = autograd.grad(u, x, torch.ones_like(u), create_graph=True)[0]

# second derivative
d2u = autograd.grad(du, x, torch.ones_like(du))[0]
```
